```python
from pyspark.sql import SparkSession

# ✅ Create Spark session
spark = SparkSession.builder \
    .appName("Chh OLA Trip Test") \
    .config("spark.master", "local[*]") \
    .getOrCreate()

# ✅ Load CSV file
df = spark.read.csv("/app/data/chh_ola_trips.csv", header=True, inferSchema=True)

# ✅ Display basic info
print("Total rows:", df.count())
df.show(5)

# ✅ Basic summary
df.describe().show()

# ✅ Stop Spark session
spark.stop()
```
